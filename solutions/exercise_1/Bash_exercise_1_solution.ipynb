## 1.8

Understand the meaning of the output in exercise 1.7. 

  * column 1: file permissions. First field indicates whether it's a file or a directory, the following fields give permissions for user, group and all in triples. r = read, w = write, x = execute
  
  * column 2: indicates how many directories reside in a directory/file with a certain name
  
  * column 3: name of user (here an odd number)
  
  * column 4: name of group (users)
  
  * column 5: file size
  
  * column 6: date of last modification
  
  * column 7: file/directory name (. is current directory, .. is the directory above)
